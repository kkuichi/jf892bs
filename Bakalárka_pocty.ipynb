```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Načítanie dát 
file_paths = {
    'wave1': "1. vlna všetko 28-11-2024.xlsx",
    'wave2': "2. vlna všetko 28-11-2024.xlsx",
    'wave3': "3. vlna všetko 28-11-2024.xlsx",
    'wave4': "4. vlna všetko 28-11-2024.xlsx"
}

data_dict = {wave: pd.read_excel(path) for wave, path in file_paths.items()}

# Funkcia na pridanie hodnôt do stĺpcových grafov
def annotate_bar(plot):
    for p in plot.patches:
        plot.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

# 2. Analýza každej vlny osobitne
def analyze_wave(wave_name, df):
    print(f"\n===== {wave_name} =====")
    print("Základné informácie:")
    print(df.info())
    print("\nPočet chýbajúcich hodnôt:")
    missing_data = df.isnull().sum()
    print(missing_data[missing_data > 0])

    # Priemerný vek a medián veku pacientov
    priemerny_vek = df["Vek"].mean()
    median_vek = df["Vek"].median()
    print(f"Priemerný vek pacientov v {wave_name}: {priemerny_vek:.2f} rokov")
    print(f"Medián veku pacientov v {wave_name}: {median_vek:.2f} rokov")

    # Distribúcia cieľovej premennej
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='Závažnosť priebehu ochorenia', data=df, palette="viridis")
    annotate_bar(ax)
    plt.xlabel("Závažnosť priebehu ochorenia")
    plt.ylabel("Počet pacientov")
    plt.title(f"Distribúcia závažnosti ochorenia - {wave_name}")
    plt.show()

    # Pohlavie
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='Pohlavie', data=df, palette="coolwarm")
    annotate_bar(ax)
    plt.xlabel("Pohlavie")
    plt.ylabel("Počet pacientov")
    plt.title(f"Pomer pohlaví - {wave_name}")
    plt.show()

    # Vek
    plt.figure(figsize=(8, 5))
    sns.histplot(df["Vek"], bins=10, kde=True, color="blue")
    plt.xlabel("Vek")
    plt.ylabel("Počet pacientov")
    plt.title(f"Distribúcia veku pacientov - {wave_name}")
    plt.show()

    # Fajčenie
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='Fajčenie', data=df, palette="coolwarm")
    annotate_bar(ax)
    plt.xlabel("Fajčiaci pacienti")
    plt.ylabel("Počet pacientov")
    plt.title(f"Pomer fajčiacich pacientov - {wave_name}")
    plt.show()

    # Alkohol
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(x='Alkohol', data=df, palette="coolwarm")
    annotate_bar(ax)
    plt.xlabel("Pacienti, ktorí požívajú alkohol")
    plt.ylabel("Počet pacientov")
    plt.title(f"Pomer pacientov závislých na alkohole - {wave_name}")
    plt.show()

    print("\nZákladné štatistiky numerických premenných:")
    print(df.describe())

# 3. Spustenie analýzy pre každú vlnu osobitne
for wave, df in data_dict.items():
    analyze_wave(wave, df)
```
